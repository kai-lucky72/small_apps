```python
import numpy as np
import matplotlib.pyplot as plt

def generate_curved_data(num_points, degree):
    # Generate random x values
    x = np.arange(num_points)
    
    # Generate y values with a polynomial curve
    coefficients = np.random.rand(degree + 1) * 10  # Random coefficients
    y = np.polyval(coefficients, x)
    
    # Add noise to y values
    noise = np.random.normal(0, 100, num_points)  # Add Gaussian noise
    y += noise
    
    return x, y

# Generate curved data points
num_points = 100
degree = 3  # Degree of polynomial curve
x, y = generate_curved_data(num_points, degree)

# Plot the generated data
plt.figure(figsize=(12, 6))
plt.scatter(x, y, label='Generated Data')
plt.title("Curved Data Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
